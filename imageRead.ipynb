```python
text_file = open('list_category_img.txt', 'r')
lines = text_file.readlines()
imgNames=lines[2:]
labels={}
categories={}
imgArray=[]
count=[]

print('289222 images; 46 categories( 3,18); 5621 labels')
for j in range(len(imgNames)):
    for i in range(len(imgNames[j])):
        if(imgNames[j][i]==' '):
            break
            
    if(imgNames[j][-3:-1] not in categories):
        categories[imgNames[j][-3:-1]]=1
    else:
        categories[imgNames[j][-3:-1]]+=1
    imgNames[j]=imgNames[j][:i]
    
i=0
label=imgNames[0][4:-17]
for imgName in imgNames:
    prev_label=label
    label=imgName[4:-17]
    if(label not in labels):
        labels[label]=[1,i,0]
        labels[prev_label][2]=i
    else:
        labels[label][0]+=1
    i+=1
    img=mpimg.imread(imgName)
    imgArray.append(img)
labels['Paisley_Maxi_Cami_Dress'][2]=len(imgNames)
print('img=mpimg.imread(imgName)')
print('imgArray.append(img)')
```
